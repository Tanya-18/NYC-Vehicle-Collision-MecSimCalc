```python
data.drop(columns=[
    'ZIP CODE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME',
    'OFF STREET NAME', 'VEHICLE 1 TYPE', 'VEHICLE 2 TYPE',
    'VEHICLE 3 TYPE', 'VEHICLE 4 TYPE', 'VEHICLE 5 TYPE',
    'VEHICLE 1 FACTOR', 'VEHICLE 2 FACTOR', 'VEHICLE 3 FACTOR',
    'VEHICLE 4 FACTOR', 'VEHICLE 5 FACTOR'], inplace=True)
data.head()
```

```python
data = data[data['LATITUDE'].notna()]
data.head()
```

```python
data[data['BOROUGH'].isna()]
```

```python
data['BOROUGH'] = data['BOROUGH'].fillna('UNSPECIFIED')
data.isna().sum()
```

```python
data.head(10)
```

```python
time = pd.DataFrame(data['TIME'])
time.head()
```

```python
time[['HOUR','MIN']] = time['TIME'].str.split(':', expand=True)
time['HOUR'] = pd.to_numeric(time['HOUR'])
time['MIN'] = pd.to_numeric(time['MIN'])
time.head(10)
```

```python
time['AM/PM'] = np.where(time['HOUR'] >= 12, "pm", "am")
time['HOUR_12'] = np.where(time['HOUR'] > 12, time['HOUR'] - 12, time['HOUR'])
time.head(10)
```

```python
time['ROUNDED'] = np.where(time['MIN'] < 30, 0, 1)
time['ROUNDED'] = np.where(time['HOUR_12'] == 12, 0, time['ROUNDED'])
time.head(10)
```

```python
time['HOUR_12_NEW'] = np.where(time['ROUNDED'] == 1, time['HOUR_12'] + 1, time['HOUR_12'])
time['HOUR_24'] = np.where(time['ROUNDED'] == 1, time['HOUR'] + 1, time['HOUR'])
time.head(10)
```

```python
time['HOUR_12_NEW'].value_counts()
```

```python
time['TIME_12'] = time['HOUR_12_NEW'].astype(str) + time['AM/PM']
time.tail(10)
```

```python
time['TIME_12'].unique()
```

```python
time['TIME_12'] = np.where(time['TIME_12'] == "0am", "12am", time['TIME_12'])
time['TIME_12'].value_counts()
```

```python
time['HOUR_24'].unique()
```

```python
time[time['HOUR_24'] == 24].head()
```

```python
time['AM/PM'] = np.where((time['HOUR'] == 23) & (time['ROUNDED'] == 1), "am", time['AM/PM'])
time['TIME_12'] = np.where((time['HOUR'] == 23) & (time['ROUNDED'] == 1), "12am", time['TIME_12'])
time['HOUR_24'] = np.where((time['HOUR'] == 23) & (time['ROUNDED'] == 1), 0, time['HOUR_24'])
time[(time['HOUR'] == 23) & (time['ROUNDED'] == 1)].head()
```

```python
time['HOUR_24'].unique()
```

```python
time.head()
```

```python
time['TOTD'] = np.where((time['HOUR_24'] >= 6) & (time['HOUR_24'] < 12), "MORNING", "")
time['TOTD'] = np.where((time['HOUR_24'] >= 12) & (time['HOUR_24'] < 18), "AFTERNOON", time['TOTD'])
time['TOTD'] = np.where((time['HOUR_24'] >= 18) & (time['HOUR_24'] < 21), "EVENING", time['TOTD'])
time['TOTD'] = np.where((time['HOUR_24'] >= 21) & (time['HOUR_24'] < 24), "NIGHT", time['TOTD'])
time['TOTD'] = np.where((time['HOUR_24'] >= 0) & (time['HOUR_24'] < 4), "LATE NIGHT", time['TOTD'])
time['TOTD'] = np.where((time['HOUR_24'] >= 4) & (time['HOUR_24'] < 6), "EARLY MORNING", time['TOTD'])
time[['TIME_12', 'HOUR_24', 'TOTD']].head(10)
```
